# Визуализация распределения переменных

**1. Гистограмма для числовой переменной Work_accident**

```python
sns.histplot(['Work_accident'], kde=True)
plt.title('Распределение Work_accident')
plt.ylabel('Частота')
plt.show()
```

**2. Столбчатая диаграмма для переменной last_evaluation**

```python
sns.countplot(x='last_evaluation', data=df)
plt.title('Распределение last_evaluation')
plt.ylabel('Количество')
plt.show()
```

3. **Боксплот для переменной satisfaction_level**

```python
sns.boxplot(x=df['satisfaction_level'])
plt.title('Boxplot satisfaction_level')
plt.xlabel('satisfaction_level')
plt.show()
```

Наблюдения: гистограмма переменной Work_accident выводится как квадрат, это может означать, что несчастные случаи бывают очень тяжкими или частыми, либо это может означать, что на этих данных невозможно анализировать.
По столбчатой диаграмме мы видим переменную last_evalution и исходя из этого можно понять, что большая половина работников не удовлетворена, а малая половина в принципе довольна.
Медиана satisfaction_level в боксплоте находится между 0.6 и 0.8.
